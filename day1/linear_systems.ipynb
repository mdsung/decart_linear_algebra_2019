# Linear Equations

Consider the equation of a line in the plane $(x_1,x_2)$.

\begin{eqnarray}
\begin{array}{rcrcr@{\qquad}l}
2x_1 & + & 4x_2 = & 2
\end{array}
\end{eqnarray}

If we plot $x_2$ vs $x_1$, we get the following plot:


**NOTE:** Our plotting function assumes that the RHS of the equation is zero. Thus the equation we need to plot is

\begin{equation}
2x_1 +4x_2-2
\end{equation}

Our plotting is based on the SymPy [plot](https://docs.sympy.org/latest/modules/plotting.html#sympy.plotting.plot.plot) function and the SymPy [solve](https://docs.sympy.org/latest/modules/solvers/solvers.html?highlight=solve#sympy.solvers.solvers.solve). With the `solvefor` keyword argument we specify which variable to plot as the dependent variable.

```python
plot_lines(2*x1+4*x2-2, solvefor=x2)
```

If we now add a second, we have a **system of linear equations**

\begin{eqnarray}
\begin{array}{rcrcrc@{\qquad}l}
2x_1 & + & 4x_2 = & 2 & (a)\\
4x_1 & + & 11x_2 = & 1 & (b)
\end{array}
\end{eqnarray}

Since we have two equations ($a$ and $b$) and two unknowns we can solve this system for $x_1$ and $x_2$:

* multiplying  $a$ by $-2$ and adding it to $b$ yields $x_2=-1$. Substituting this back into $a$ yields $x_1=3$.

If we plot these two equations, the solution (if it exists) is where the two lines intersect.

```python
plot_lines(2*x1+4*x2-2, 4*x1+11*x2-1, solvefor=x2)
```

If we change the equation slightly to

\begin{eqnarray}

4x_1 & + & 8x_2 = & 1 & (b)
\end{array}
\end{eqnarray}
